```
!pip install extract_msg
import pandas as pd
import sys
import extract_msg
import glob  

label = []
content = []

# Create a dataframe contaning files names and categories of emails

df = pd.DataFrame()


path = './dataset/MDU/*.msg'   
files=glob.glob(path) 
print(len(files))

for file in files:     
    f=open(file, 'r')  
    msg = extract_msg.Message(file)
    msg_subj = msg.subject
    msg_body = msg.body
    label.append("mdu")
    content.append(msg_subj + msg_body)

path = './dataset/Retirements/*.msg'   
files=glob.glob(path) 
print(len(files))

for file in files:     
    f=open(file, 'r')  
    msg = extract_msg.Message(file)
    msg_subj = msg.subject
    msg_body = msg.body
    label.append("retirements")
    content.append(msg_subj + msg_body)


path = './dataset/Transfers_2/*.msg'   
files=glob.glob(path) 
print(len(files))

for file in files:     
    f=open(file, 'r')  
    msg = extract_msg.Message(file)
    msg_subj = msg.subject
    msg_body = msg.body
    label.append("transfers")
    content.append(msg_subj + msg_body)

path = './dataset/Transfer/*.msg'   
files=glob.glob(path) 
print(len(files))

for file in files:     
    f=open(file, 'r')  
    msg = extract_msg.Message(file)
    msg_subj = msg.subject
    msg_body = msg.body
    label.append("transfers")
    content.append(msg_subj + msg_body)
  

path = './dataset/Retirement_2/*.msg'   
files=glob.glob(path) 
print(len(files))

for file in files:     
    f=open(file, 'r')  
    msg = extract_msg.Message(file)
    msg_subj = msg.subject
    msg_body = msg.body
    label.append("retirements")
    content.append(msg_subj + msg_body)




path = './dataset/MDU_2/*.msg'   
files=glob.glob(path) 
print(len(files))

for file in files:     
    f=open(file, 'r')  
    msg = extract_msg.Message(file)
    msg_subj = msg.subject
    msg_body = msg.body
    label.append("mdu")
    content.append(msg_subj + msg_body)



df["content"] = content
df["label"] = label
df = df.sample(frac=1).reset_index(drop=True)
df.to_csv('./files/train_3.csv')
df.head()
```

```
#finding frequencies of different words in the dataset before cleaning
key_words={}

df1 = df[df['label']=='retirements']

l1 = df1['content']


for i in l1:
  arr = i.split(' ')
  for word in arr:
    if word in key_words.keys():
      key_words[word]+=1
    else:
      key_words[word]=1


sort_orders = sorted(key_words.items(), key=lambda x: x[1], reverse=True)   

for j in sort_orders:
  print(j)

```
